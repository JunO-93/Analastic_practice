##데이터를 클래스별로 타겟하여 마스킹

```
class0=nd_wines_data[nd_wines_target==0]
class1=nd_wines_data[nd_wines_target==1]
class2=nd_wines_data[nd_wines_target==2]
```

```
len(class0),len(class1),len(class2)
```

##class0의 해당 하는 열의 평균,중앙값,표준편차,분산

```
class0_mean=[]
class0_median=[]
class0_std=[]
class0_var=[]
class0_avg=[]

for i in range(12):
  class0_avg +=[round(np.mean(class0[:,i]),2)] 
  class0_mean +=[round(np.mean(class0[:,i]),2)]  
  class0_median += [np.median(class0[:,i])]
  class0_std += [round(np.std(class0[:,i]),2)]
  class0_var += [round(np.var(class0[:,i]),2)]
  print(i,"번째 min 값 :",np.min(class0[:,i]))
  print(i,"번째 max 값 :",np.max(class0[:,i]))
  print("-------------------")
```

```
print("평균:     ",class0_mean)
print("중앙값:   ",class0_median)
print("표준편차: ",class0_std)
print("분산:     " ,class0_var)
```

- class0에 해당된 컬럼들의 min,max의 극단치의 차이가 얼마나지 않고, 표준편차도 적은 값으로 판단되어 대푯값은 평균을 사용해도 된다고 생각합니다.

- Magnesium의 극단치가 차이가 있지만 평균과 중앙값을 비교했을 때 차이가 2.34으로 적다고 판단되어 똑같이 대푯값은 평균으로 사용하면 좋을 것 같습니다.



```
0#매트릭스로 정리
class0_mean=np.insert(class0_mean,12,class0_median,axis=0)
class0_mean=np.insert(class0_mean,24,class0_std,axis=0)
class0_mean=np.insert(class0_mean,36,class0_var,axis=0)

#4*12 매트릭스로 리사이즈
class0_mean=np.resize(class0_mean,(4,12))
class0_mean
```
