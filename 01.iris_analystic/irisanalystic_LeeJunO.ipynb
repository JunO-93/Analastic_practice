# ● 데이터시각화(산점도)

```
nd_iris_data= np.array(iris['data'])
nd_iris_target = np.array(iris['target']) #타켓
type(nd_iris_data)
```

```
sepal_length= nd_iris_data[:,0] #sepal length (cm)
sepal_width= nd_iris_data[:,1]  #sepal width (cm)
petal_length= nd_iris_data[:,2] #petal length (cm)
petal_width= nd_iris_data[:,3]  #petal width (cm)
```

##Setosa

```
setosa_sep_length = sepal_length[nd_iris_target==0]  #sepal length masking
setosa_sep_width = sepal_width[nd_iris_target==0]  #sepal width masking

setosa_pet_length = petal_length[nd_iris_target==0]  #petal length masking
setosa_pet_width = petal_width[nd_iris_target==0]  #petal width masking


print(setosa_sep_length)
print(setosa_sep_width)
print(setosa_pet_length)
print(setosa_pet_width)
```

- setosa의 sepal과 petal 산점도

```
plt.scatter(setosa_sep_length,setosa_sep_width,marker='v',color='g')
plt.scatter(setosa_pet_length,setosa_pet_width,marker='^',color='m')

plt.title('setosa sepal,pet length and width')
plt.show()
```

##Versicolour

```
versicolour_sep_length = sepal_length[nd_iris_target==1]  #sepal length masking
versicolour_sep_width = sepal_width[nd_iris_target==1]  #sepal width masking

versicolour_pet_length = petal_length[nd_iris_target==1]  #petal length masking
versicolour_pet_width = petal_width[nd_iris_target==1]  #petal width masking


print(versicolour_sep_length)
print(versicolour_sep_width)
print(versicolour_pet_length)
print(versicolour_pet_width)
```

- Versicolour의 sepal과 petal 군집도

```
plt.scatter(versicolour_sep_length,versicolour_sep_width,marker='v',color='g')
plt.scatter(versicolour_pet_length,versicolour_pet_width,marker='^',color='m')

plt.title('Versicolour sepal,pet length and width')
plt.show()
```

##Virginica

```
virginicar_sep_length = sepal_length[nd_iris_target==2]  #sepal length masking
virginicar_sep_width = sepal_width[nd_iris_target==2]  #sepal width masking

virginicar_pet_length = petal_length[nd_iris_target==2]  #petal length masking
virginicar_pet_width = petal_width[nd_iris_target==2]  #petal width masking


print(virginicar_sep_length)
print(virginicar_sep_width)
print(virginicar_pet_length)
print(virginicar_pet_width)
```

- Virginica의 sepal과 petal 산점도

```
plt.scatter(virginicar_sep_length,virginicar_sep_width,marker='v',color='g')
plt.scatter(virginicar_pet_length,virginicar_pet_width,marker='^',color='m')

plt.title('Virginica sepal,pet length and width')
plt.show()
```

## 종합 산점도

```
plt.scatter(setosa_sep_length,setosa_sep_width,marker='^',color='r')
plt.scatter(setosa_pet_length,setosa_pet_width,marker='^',color='g')
plt.scatter(versicolour_sep_length,versicolour_sep_width,marker='^',color='b')
plt.scatter(versicolour_pet_length,versicolour_pet_width,marker='^',color='y')
plt.scatter(virginicar_sep_length,virginicar_sep_width,marker='^',color='m')
plt.scatter(virginicar_pet_length,virginicar_pet_width,marker='^',color='c')

plt.legend(['set_sep','set_pet','ver_sep','ver_pet','vir_sep','vir_pet'])
plt.title('Petal length,width and Sepal length,width by iris class')
plt.show()
```

##아이리스 종류별 산점도

```
plt.scatter(setosa_sep_length,setosa_sep_width,marker='v',color='g')
plt.scatter(versicolour_sep_length,versicolour_sep_width,marker='^',color='b')
plt.scatter(virginicar_sep_length,virginicar_sep_width,marker='^',color='m')

plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```
plt.scatter(setosa_pet_length,setosa_pet_width,marker='^',color='g')
plt.scatter(versicolour_pet_length,versicolour_pet_width,marker='^',color='y')
plt.scatter(virginicar_pet_length,virginicar_pet_width,marker='^',color='c')

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```
