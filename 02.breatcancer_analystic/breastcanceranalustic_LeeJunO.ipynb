#시각화

- malignant : 악성 #212개, 0
- benign : 양성 #357개, 1

```
mal= ndCancer[cancer['target']==0]
ben= ndCancer[cancer['target']==1]
```

```
#radius #20 
#texture #21
#perimeter #22
#area #23
```

```
# 악성,양성을 반지름,텍스처,둘레,영역별로 구분
mal_rad = mal[:,20]
mal_tex = mal[:,21]
mal_per = mal[:,22]
mal_area = mal[:,23]

ben_rad = ben[:,20]
ben_tex = ben[:,21]
ben_per = ben[:,22]
ben_area = ben[:,23]
```

```
len(mal_rad)
```

```
len(ben_rad)
```

- 길이와 텍스처를 악성과 양성으로 비교했을때, 구별이 가능한지 여부를 알 수 있다.

```
plt.scatter(mal_rad,mal_tex,color='m')
plt.scatter(ben_rad,ben_tex,color='c')
plt.xlabel('radius')
plt.ylabel('texture')
plt.legend(['malignant','benign'])
plt.show
```


- 둘레와 영역으로 악성,양성을 구별이 가능한지 여부를 알 수 있다.

```
plt.scatter(mal_per,mal_area,color='r')
plt.scatter(ben_per,ben_area,color='b')
plt.xlabel('perimeter')
plt.ylabel('area')
plt.legend(['malignant','benign'])
plt.show
```
