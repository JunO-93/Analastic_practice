- CRIM: 범죄율

- RM: 주택당 방 수

- B: 인구 중 흑인 비율

- ZN: 25,000 평방피트를 초과 거주지역 비율

- AGE: 1940년 이전에 건축된 주택의 비율
- RAD: 방사형 고속도로까지의 거리
- DIS: 직업센터의 거리
- TAX: 재산세율
- MEDV   보스턴 506개 타운의 1978년 주택 가격 중앙값 (단위 1,000 달러)

```
nd_boston_data = np.array(boston['data'],dtype= np.int64)
nd_boston_target = np.array(boston['target'])
```

```
crim =[nd_boston_data[:,0]]
zn = [nd_boston_data[:,1]]
indus = [nd_boston_data[:,2]]
chas = [nd_boston_data[:,3]]
nox = [nd_boston_data[:,4]]
rm = [nd_boston_data[:,5]]
age = [nd_boston_data[:,6]]
dis = [nd_boston_data[:,7]]
rad = [nd_boston_data[:,8]]
tax = [nd_boston_data[:,9]]
ptratio = [nd_boston_data[:,10]]
b = [nd_boston_data[:,11]]
lstat = [nd_boston_data[:,12]]
```

```
print(nd_boston_data)
```

```
boston_mean =[]  
boston_median = []
boston_std = []
boston_var=[]

for i in range(13):
  boston_mean +=[round(np.mean(nd_boston_data[:,i]),2)]  
  boston_median += [np.median(nd_boston_data[:,i])]
  boston_std += [round(np.std(nd_boston_data[:,i]),2)]
  boston_var += [round(np.var(nd_boston_data[:,i]),2)]
  print(i,"번째 min 값 :",np.min(nd_boston_data[:,i]))
  print(i,"번째 max 값 :",np.max(nd_boston_data[:,i]))
  print()
```

```
print(boston_mean)
print(boston_median)
print(boston_std)
print(boston_var)
```

- CRIM,ZN,INDUS,AGE,TAX ,B 컬럼의 극단치는 높으므로 대푯값으로 중앙값을 사용하는것이 좋다고 판단됩니다.
- CHAS,NOX,DIS,RAD,PTRATIO,LSTAT 컬럼의 극단치는 낮으므로 대푯값으로 평균을 사용하는게 좋다고 판단됩니다.
